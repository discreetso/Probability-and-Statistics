```python
# data visualization
```

```python
# prepare some data
y = [6, 7, 2, 4, 5]
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(y)         # this will plot a line graph
plt.show()
```

```python
# create a figure 
plt.figure(figsize = (10, 5))

# add a line renderer with legend and line thickness
plt.plot(y)

# also need to make the figure readable -- very important from stats perspective
plt.title("First example plot")
plt.xlabel("x vals (s)")
plt.ylabel("y vals (m)")
plt.grid()
plt.show()
```

```python
# plotting weights
# create a figure 
plt.figure(figsize = (20, 5))

# add a line renderer with legend and line thickness
x = range(da["BMXWT"].size)
y = da["BMXWT"]
plt.scatter(x, y, marker='x', s =0.5)

# also need to make the figure readable -- very important from stats perspective
plt.title("Weight of Individuals")
plt.xlabel("ID of individuals")
plt.ylabel("Weight (kg)")
plt.show()
```
