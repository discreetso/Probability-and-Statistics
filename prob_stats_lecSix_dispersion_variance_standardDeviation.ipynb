# Generating a lot of values

```python
num_samples = 10000          # get rid of 'magic numbers'
```

```python
uniform_vals = np.random.uniform(low=0.0, high=10.0, size=num_samples)
```

```python
sns.displot(uniform_vals, bins=20, kde=False)
plt.ylabel('Frequency')
plt.xlabel('values')
plt.title('uniform values')
sns.despine(offset=10, trim=True)        # move axes away
plt.show()
```

```python
normal_vals = np.random.normal(loc=5.0, scale=10.0, size=num_samples)
```

```python
sns.displot(normal_vals, bins=20, kde=False)
plt.xlabel('values')
plt.ylabel('frequency')
plt.title('normal values')
sns.despine(offset=10, trim=True)       # move axes away
plt.show()
```

```python
print('Uniform values mean: ', np.mean(uniform_vals))
print('Normal values mean: ', np.mean(normal_vals))
```

```python
print('Uniform values variance: ', np.var(uniform_vals))
print('Normal values variance: ', np.var(normal_vals))
```

```python
print('Uniform values standard deviation(sd): ', np.std(uniform_vals))
print('Normal values standard deviation(sd): ', np.std(normal_vals))
```

```python
normal_vals = np.random.normal(loc=5.0, scale=10.0, size=num_samples)
nv_mean = np.mean(normal_vals)
nv_sd = np.std(normal_vals)
```

```python
sns.displot(normal_vals, bins=20, kde=False)
plt.xlim(-60, 60)
plt.xlabel('values')
plt.ylabel('frequency')
plt.title('Normal values')
sns.despine(offset=10, trim=True)           # move axes away

# plot the mean line
x_c, y_c = ([nv_mean, nv_mean], [0, 1500])
plt.plot(x_c, y_c, color='red', linewidth=2)

# plot the SD line on right side of the mean
x_c, y_c = ([nv_mean + nv_sd, nv_mean + nv_sd], [0, 1500])
plt.plot(x_c, y_c, color='green', linewidth=2)

# plot the SD line on left side of the mean
x_c, y_c = ([nv_mean - nv_sd, nv_mean - nv_sd], [0, 1500])
plt.plot(x_c, y_c, color='green', linewidth=2)

plt.show()
```

```python
# plotting the above plot again but only changing its scale to see the difference, no difference will be seen if no knowledge but there's a clear
# difference in the units which will be unseen by the clients, we have to made it clear by setting the both plots limits same 
normal_vals_2 = np.random.normal(loc=5.0, scale=10.0, size=num_samples)
nv_mean_2 = np.mean(normal_vals)
nv_sd_2 = np.std(normal_vals)
```

```python
sns.displot(normal_vals_2, bins=20, kde=False)
plt.xlim(-60, 60)
plt.xlabel('values')
plt.ylabel('frequency')
plt.title('Normal values 2')
sns.despine(offset=10, trim=True)           # move axes away

# plot the mean line
x_c, y_c = ([nv_mean_2, nv_mean_2], [0, 1500])
plt.plot(x_c, y_c, color='red', linewidth=2)

# plot the SD line on right side of the mean
x_c, y_c = ([nv_mean_2 + nv_sd_2, nv_mean_2 + nv_sd_2], [0, 1500])
plt.plot(x_c, y_c, color='green', linewidth=2)

# plot the SD line on left side of the mean
x_c, y_c = ([nv_mean_2 - nv_sd_2, nv_mean_2 - nv_sd_2], [0, 1500])
plt.plot(x_c, y_c, color='green', linewidth=2)

plt.show()
```

```python
med_incs = [x[0] for x in houses.data]
mean = np.mean(med_incs)
sd = np.std(med_incs)

sns.displot(med_incs, bins=20, kde=False)
plt.xlim(0, 16)
plt.xlabel('Frequency')
plt.ylabel('med_incs')
plt.title('Histogram of med_incs')
sns.despine(offset=10, trim=True)

# plot the mean line
x_c, y_c = ([mean, mean], [0, 4000])
plt.plot(x_c, y_c, color='green', linewidth=2)

# plot the SD right to the mean
x_c, y_c = ([mean + sd, mean + sd], [0, 4000])
plt.plot(x_c, y_c, color='red', linewidth=2)

# plot the SD right to the mean
x_c, y_c = ([mean - sd, mean - sd], [0, 4000])
plt.plot(x_c, y_c, color='red', linewidth=2)
plt.show()
```
